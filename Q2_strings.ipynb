### Exercise 1 - First, last, length  ·  *easy*

Print the first character, the last character, and the length of the string.

### Exercise 8 - Number in a sentence  ·  *medium*

The line below breaks. Fix it two different ways.

````python
year = 1982

# This is broken - you cannot add a number to a string.
# print("Released in " + year)

# Fix 1: cast the number to a string.
print("Released in " + str(year))

# Fix 2: give print() several things and let it handle the joining.
print("Released in", year)

# Fix 3 (the modern way): an f-string. Put the variable in curly braces.
print(f"Released in {year}")
````

### Exercise 10 - Clean up a messy name  ·  *hard*

A user typed their name badly. Clean it into `Asha Verma`.

````python
messy = "   aSHa   verma  "

# Step 1: remove the spaces at both ends.
# Step 2: fix the capitalisation.
# Step 3: squeeze the double space in the middle down to one.

clean = messy.strip() # Remove leading/trailing spaces

clean = clean.replace('  ', ' ') # Squeeze multiple spaces into single spaces using replace

clean = clean.title() # Apply title case

print(f"[{clean}]") # Expect: [Asha Verma]
````

### Exercise 11 - Split a record  ·  *hard*

A line of data arrived as one string. Pull out the three pieces.

````python
row = "Asha,Delhi,88"

# split() cuts a string into a list, wherever it finds the separator.
parts = row.split(",")
print(parts)          # Expect: ['Asha', 'Delhi', '88']

name = parts[0]
city = parts[1]
marks = int(parts[2])   # this must be a NUMBER, not text

print(f"{name} from {city} scored {marks}")
print(marks + 2)      # Expect: 90, not '882'
````

### Exercise 12 - Is it a palindrome?  ·  *hard*

A palindrome reads the same forwards and backwards, ignoring case and spaces.
Write one expression that answers it.

````python
word = "Malayalam"

# Hint: lower it, then compare it to its reverse.
print(word.lower()==word[::-1].lower())       # Expect: True

word = "Python"
print(word.lower()==word[::-1].lower())       # Expect: False
````
