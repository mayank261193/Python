### Exercise 4 - Cut off, not rounded  ·  *easy*

`int()` throws the decimal away. `round()` rounds properly.
Print both for the value `4.7` and see the difference.

```python
value = 4.7

print(int(value))     # 4  - int() cuts the decimal off
print(round(value))   # 5  - round() rounds to the nearest whole number
```

### Exercise 5 - The four divisions  ·  *medium*

Python has more than one kind of division. Print all four results for 17 and 5.

```python
a = 17
b = 5

print(a / b)     # 3.4      - always returns a float
print(a // b)    # 3        - integer division, rounds down
print(a % b)     # 2        - the remainder
print(a ** b)    # 1419857  - power
```

### Exercise 9 - Was it from the 80s?  ·  *hard*

Write one expression that is `True` only if `album_year` is between 1980 and 1989 inclusive.

```python
album_year = 1982
print(album_year >= 1980 and album_year <= 1989)   # True

album_year = 1991
print(album_year >= 1980 and album_year <= 1989)   # False

# Python also allows this shorter form, which reads like maths:
album_year = 1982
print(1980 <= album_year <= 1989)                  # True
```

### Exercise 12 - Your first tiny program  ·  *hard*

A shop sells tea at 10 rupees and coffee at 25 rupees.
A customer buys 3 teas and 2 coffees. There is 5% GST on the total.

Print the final amount, rounded to 2 decimal places.
Use variables with clear names - do not just write one long sum.

```python
tea_price = 10
coffee_price = 25

tea_qty = 3
coffee_qty = 2

subtotal = (tea_price * tea_qty) + (coffee_price * coffee_qty)   # 80
gst = subtotal * 0.05                                            # 4.0
final = subtotal + gst                                           # 84.0

print("Final amount:", round(final, 2))

# Why variables and not one long line?
# Because in two months this becomes 40 lines, and you will need to read it.
```
