### Exercise 12 - Your first tiny program  ·  *hard*

A shop sells tea at 10 rupees and coffee at 25 rupees.
A customer buys 3 teas and 2 coffees. There is 5% GST on the total.

Print the final amount, rounded to 2 decimal places.
Use variables with clear names - do not just write one long sum.

```python
# YOUR CODE HERE
# Expected output: Final amount: 84.0
Tea_cost=10
Coffee_cost=25
GST=0.05
Tea_amount=3
Coffee_amount=2
Total_amount=(Tea_cost*Tea_amount)+(Coffee_cost*Coffee_amount)
GST_amount=Total_amount*GST
Final_amount=Total_amount+GST_amount
print("Final amount:", round(Final_amount,2))
```
